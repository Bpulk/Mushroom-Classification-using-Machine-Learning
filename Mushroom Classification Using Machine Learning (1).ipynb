# Applying PCA (principle component analysis)

# splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(pca_fit,y,test_size=0.20,random_state=42)
```

#importing the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

# model training

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)

knn = KNeighborsClassifier()
knn.fit(x_train,y_train)

svc = SVC()
svc.fit(x_train,y_train)

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

rm = RandomForestClassifier()
rm.fit(x_train,y_train)

gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)

xgb = XGBClassifier()
xgb.fit(x_train,y_train)

ada = AdaBoostClassifier()
ada.fit(x_train,y_train)

cat = CatBoostClassifier()
cat.fit(x_train,y_train)

mlp = MLPClassifier()
mlp.fit(x_train,y_train)
```

# prediction on test data

```python
y_pred1 = lr.predict(x_test)
y_pred2 = knn.predict(x_test)
y_pred3 = svc.predict(x_test)
y_pred4 = dt.predict(x_test)
y_pred5 = rm.predict(x_test)
y_pred6 = gb.predict(x_test)
y_pred7 = xgb.predict(x_test)
y_pred8 = ada.predict(x_test)
y_pred9 = cat.predict(x_test)
y_pred10 = mlp.predict(x_test)
```

# evaluating the algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
print ("ACC LR", accuracy_score(y_test,y_pred1))
print ("ACC KNN", accuracy_score(y_test,y_pred2))
print ("ACC SVC", accuracy_score(y_test,y_pred3))
print ("ACC DT", accuracy_score(y_test,y_pred4))
print ("ACC RM", accuracy_score(y_test,y_pred5))
print ("ACC GB", accuracy_score(y_test,y_pred6))
print ("ACC XGB", accuracy_score(y_test,y_pred7))
print ("ACC ADA", accuracy_score(y_test,y_pred8))
print ("ACC CAT", accuracy_score(y_test,y_pred9))
print ("ACC MLP", accuracy_score(y_test,y_pred10))
```

```python
final_data = pd.DataFrame({'Models':['LR','KNN','SVC','DT','RM','GB','XGB','ADA','CAT','MLP'],
              'ACC' : [accuracy_score(y_test,y_pred1)*100,
                       accuracy_score(y_test,y_pred2)*100,
                       accuracy_score(y_test,y_pred3)*100,
                       accuracy_score(y_test,y_pred4)*100,
                       accuracy_score(y_test,y_pred5)*100,
                       accuracy_score(y_test,y_pred6)*100,
                       accuracy_score(y_test,y_pred7)*100,
                       accuracy_score(y_test,y_pred8)*100,
                       accuracy_score(y_test,y_pred9)*100,
                       accuracy_score(y_test,y_pred10)*100]})
```

# Classification Report for all the models

```python
from sklearn.metrics import classification_report
model_names = ['LR', 'KNN', 'SVC', 'DT', 'RM', 'GB','XGB','ADA','CAT','MLP']
predictions = [y_pred1, y_pred2, y_pred3, y_pred4, y_pred5, y_pred6, y_pred7, y_pred8, y_pred9, y_pred10]
classification_reports = {}
for model_name, y_pred in zip(model_names, predictions):
    report = classification_report(y_test, y_pred)
    classification_reports[model_name] = report
for model_name, report in classification_reports.items():
    print(f"Classification Report for {model_name}:")
    print(report)
```

# confusion matrix heatmap for each of the model

```python
from sklearn.metrics import confusion_matrix
model_names = ['LR', 'KNN', 'SVC', 'DT', 'RM', 'GB', 'XGBoost', 'AdaBoost', 'CatBoost', 'MLP']
classifiers = [lr, knn, svc, dt, rm, gb, xgb, ada, cat, mlp]
fig, axes = plt.subplots(5, 2, figsize=(12, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

color_palettes = ['Blues', 'Reds', 'Greens', 'Oranges', 'Purples', 'YlOrBr', 'coolwarm', 'RdYlBu', 'viridis', 'cividis']

for i, (model_name, classifier, color_palette) in enumerate(zip(model_names, classifiers, color_palettes)):
    row, col = divmod(i, 2)
    ax = axes[row, col]

    y_pred = classifier.predict(x_test)
    cm = confusion_matrix(y_test, y_pred)

    sns.heatmap(cm, annot=True, fmt='d', cmap=color_palette, ax=ax)

    ax.set(title=f'Confusion Matrix - {model_name}',
           xlabel='Predicted',
           ylabel='Actual')

plt.show()
```

# ROC curves of each model

```python

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, auc

# Create a list of model names and their corresponding predicted probabilities
model_names = ['LR', 'KNN', 'SVC', 'DT', 'RM', 'GB','XGB','ADA','CAT','MLP']
predicted_probabilities = [lr.predict_proba(x_test)[:, 1], knn.predict_proba(x_test)[:, 1],
                           svc.decision_function(x_test), dt.predict_proba(x_test)[:, 1],
                           rm.predict_proba(x_test)[:, 1], gb.predict_proba(x_test)[:, 1],xgb.predict_proba(x_test)[:, 1],
                           ada.predict_proba(x_test)[:, 1],cat.predict_proba(x_test)[:, 1],mlp.predict_proba(x_test)[:, 1]]

# Plot ROC curves for each model
plt.figure(figsize=(10, 6))
for model_name, y_prob in zip(model_names, predicted_probabilities):
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc='lower right')
plt.show()


```

# class prediction error plots for each of the model

```python
import numpy as np
model_names = ['LR', 'KNN', 'SVC', 'DT', 'RM', 'GB', 'XGBoost', 'AdaBoost', 'CatBoost', 'MLP']
classifiers = [lr, knn, svc, dt, rm, gb, xgb, ada, cat, mlp]
fig, axes = plt.subplots(5, 2, figsize=(12, 20))
plt.subplots_adjust(wspace=0.5, hspace=0.5)
for i, (model_name, classifier) in enumerate(zip(model_names, classifiers)):
    row, col = divmod(i, 2)
    ax = axes[row, col]

    y_pred = classifier.predict(x_test)
    cm = confusion_matrix(y_test, y_pred)

    class_prediction_errors = np.diag(cm) / np.sum(cm, axis=1)

    ax.bar(np.arange(len(class_prediction_errors)), class_prediction_errors, align='center')

    ax.set(xticks=np.arange(2),
           xticklabels=['Class 0', 'Class 1'],
           title=f'Class Prediction Error - {model_name}',
           ylabel='Prediction Error',
           ylim=(0, 1))

plt.show()


```

# Learning curve, validation curve of random forest classifier

```python
import numpy as np
from sklearn.model_selection import learning_curve, validation_curve
```

```python
rf_classifier = RandomForestClassifier()
rf_classifier.fit(x_train, y_train)

# Learning Curve
train_sizes, train_scores, test_scores = learning_curve(rf_classifier, x_train, y_train, cv=5, scoring='accuracy')

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend()
plt.show()

# Validation Curve
param_range = [10, 50, 100, 200]
train_scores, test_scores = validation_curve(
    RandomForestClassifier(), x_train, y_train, param_name="n_estimators", param_range=param_range, cv=5, scoring="accuracy"
)

plt.figure(figsize=(8, 6))
plt.plot(param_range, np.mean(train_scores, axis=1), label='Train')
plt.plot(param_range, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.title('Validation Curve for Number of Estimators')
plt.legend()
plt.show()
```
